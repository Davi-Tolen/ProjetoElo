```python
# Importando DemographicData_ZCTAs

import pandas as pd
import numpy as np

DemographicData_ZCTAs = pd.read_csv('../bases_iniciais/DemographicData_ZCTAs.csv')
DemographicData_ZCTAs
```

```python
# Deletando coluna "Unnamed: 0"

DemographicData_ZCTAs = DemographicData_ZCTAs.drop(columns=['Unnamed: 0'])
DemographicData_ZCTAs
```

```python
# checando valores nulos

DemographicData_ZCTAs.isna().sum()
```

```python
DemographicData_ZCTAs[DemographicData_ZCTAs['SexRatio(males per 100 females)'].isna()]
```

```python
# dados com valores irrelevantes, faz mais sentido deletar as linhas

DemographicData_ZCTAs = DemographicData_ZCTAs.dropna()
DemographicData_ZCTAs
```

```python
# resetando index

DemographicData_ZCTAs.reset_index(inplace=True, drop=True)
DemographicData_ZCTAs
```

```python
# checando se os valores nulos foram deletados

DemographicData_ZCTAs.isna().sum()
```

```python
# checando valores duplicados

DemographicData_ZCTAs.duplicated().sum()
```

```python
# conferindo

DemographicData_ZCTAs['Id'].duplicated().sum()
```

```python
# tornando nomes como padrão python

DemographicData_ZCTAs = DemographicData_ZCTAs.rename(columns = {'Id':'id', 'GeographicAreaName':'geographic_area_name', 'TotalPopulation': 'total_population' })
DemographicData_ZCTAs
```

```python
# salvando e exportando base transformada

DemographicData_ZCTAs.to_csv('../bases_finais/final_DemographicData_ZCTAs.csv', sep=';', index=False)
```
