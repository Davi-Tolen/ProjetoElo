```python
# Importando df_geocode

import pandas as pd
import numpy as np

df_geocode = pd.read_csv('../bases_iniciais/df_geocode.csv')
df_geocode
```

```python
# Deletando coluna "Unnamed: 0"

df_geocode = df_geocode.drop(columns=['Unnamed: 0'])
df_geocode
```

```python
# checando valores nulos

df_geocode.isna().sum()
```

```python
# checando dado com valor nulo

df_geocode[df_geocode['Zipcode'].isna()]
```

```python
# como o dado acima nÃ£o tem sentido, vou retirar ele da base de dados

df_geocode = df_geocode.dropna()
df_geocode
```

```python
# checando se a linha foi excluÃ­da

df_geocode.isna().sum()
```

```python
# resetando index

df_geocode.reset_index(inplace=True, drop=True)
df_geocode
```

```python
# checando valores duplicados

df_geocode.duplicated().sum()
```

```python
# sÃ³ para ter certeza

df_geocode['Lab Id'].duplicated().sum()
```

```python
# tornando nomes como padrÃ£o python

df_geocode = df_geocode.rename(columns = {'Lab Id':'lab_id', 'Address':'address', 'Location':'location', 'Zipcode': 'zipcode'})
```

```python
# transformando zipcode em int e nÃ£o float

df_geocode[['zipcode']] = df_geocode[['zipcode']].astype('int64')
df_geocode
```

```python
# avaliando incoerÃªncias

df_geocode.describe(include='all')
```
